```python
import requests
```

```python
import datetime
```

```python
from bs4 import BeautifulSoup
```

```python
url ='https://www.ccny.cuny.edu/registrar/fall'
```

```python
page = requests.get(url)
```

```python
import pandas as pd
```

```python
soup = BeautifulSoup(page.content, 'html')
```

```python
table = soup.find_all('table')[0]
```

```python
print(table)
```

```python
#world_titles = table.find_all('p')
items = [p.get_text(strip=True) for p in soup.find("table").find_all("td")]
```

```python
print(items)
```

```python
rows = []
```

```python
for i in range(0, len(items)-2, 3):
    date_str, dow, text = items[i], items[i+1], items[i+2]


    if "-" in date_str:
        continue

    try:
        date = datetime.datetime.strptime(date_str + " 2021", "%B %d %Y").date()
    except ValueError:
        continue

    rows.append([date, dow, text])
```

```python
# DataFrame
df = pd.DataFrame(rows, columns=["date", "dow", "text with the explanation"]).set_index("date")

df
```
